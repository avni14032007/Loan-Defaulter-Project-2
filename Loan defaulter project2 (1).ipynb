```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\credit_risk_dataset.csv")
```

```python
a=df.isnull().sum()/df.shape[0]*100
print(a)
```

```python
df_num_fin=df.select_dtypes(include=['int','float'])
print(df_num_fin.isnull().sum())
```

```python
df_num_fin['Income'].replace(np.nan, df_num_fin['Income'].median(), inplace=True)
```

```python
df_obj_fin=df.select_dtypes(include='object')
```

```python
df_obj_fin=pd.get_dummies(df_obj_fin,dtype=int,drop_first=True)
```

```python
df_new=pd.concat([df_num_fin,df_obj_fin],axis=1)
```

```python
X=df_new.drop(columns='Default')
y=df_new['Default']
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test=train_test_split(X,y,test_size=.2, random_state=12,shuffle=True,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ar=ss.fit_transform(X_train).round(1)
X_train_scl=pd.DataFrame(ar,columns=X_train.columns)
```

```python
print(X_train_scl.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ar=ss.fit_transform(X_test).round(1)
X_test_scl=pd.DataFrame(ar,columns=X_test.columns)
```

```python
print(X_test_scl.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_scl,y_train)
y_pred=lr.predict(X_test_scl)
```

```python
print(lr.score(X_test_scl,y_test))
```
